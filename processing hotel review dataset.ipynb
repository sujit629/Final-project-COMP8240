```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = r"C:\Users\Sujit\Downloads\Datafiniti_Hotel_Reviews\Datafiniti_Hotel_Reviews.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

```

```python
df = df.drop(columns=['id','dateAdded','dateUpdated','address','categories','primaryCategories','city','country','keys','latitude','longitude','name','postalCode','province','reviews.date','reviews.dateSeen','reviews.sourceURLs','reviews.title','reviews.userCity','reviews.userProvince','reviews.username','sourceURLs','websites'])
```

```python
df.head()
```

```python
# Function to set labels for review ratings
def set_label(review_rating):
    if 1.00 <= review_rating <= 2.50:
        return 0
    elif 2.50 < review_rating <= 5.00:
        return 1
    else:
        return None  
```

```python
# Apply the set_label function to the 'reviews.rating' column and create a new 'label' column
df['label'] = df['reviews.rating'].apply(set_label)

# Print the DataFrame to see the new 'label' column
print(df)
```
